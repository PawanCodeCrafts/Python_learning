```python
n = 77
```

```python
if n % 2 == 0:
    if n % 3 == 0:
        if n % 5  == 0:
            print('divisible by 2 , 3, 5')
        else:
            print('divisible by 2, 3, !5')
    elif n % 5 == 0:
        print('divisible by 2, 5, !3')
    else:
        print('divisible by 2 , !3, !5')

elif n % 3 == 0:
    if n % 5 == 0:
        print('divisible by 3 , 5')
    else:
        print('divisible by 3 , !5 , !2')

elif n % 5 == 0:
    print('divisible by 5')
    
else:
    print('divisible by !2, !3, !5')
```
